Esto son los metadatos

```python
# Id: Unique identifier for each property.
# 1- PassengerId - A unique Id for each passenger. 
# Each Id takes the form gggg_pp where gggg indicates a group the passenger is travelling with and pp is their number within the group. 
# People in a group are often family members, but not always.
# ----------------------
# 2- HomePlanet - The planet the passenger departed from, typically their planet of permanent residence.
# ----------------------
# 3- CryoSleep - Indicates whether the passenger elected to be put into suspended animation for the duration of the voyage.
# Passengers in cryosleep are confined to their cabins.
# ----------------------
# 4- Cabin - The cabin number where the passenger is staying. 
# Takes the form 
#     deck/num/side, 
#     deck = cubierta barco 
#     num = numero_cabina
#     where side can be either P for Port or S for Starboard. (por babor o por estribor)
# ----------------------
# 5- Destination - The planet the passenger will be debarking to.
# ----------------------
# 6- Age - The age of the passenger.
# ----------------------
# 7- VIP - Whether the passenger has paid for special VIP service during the voyage. RoomService, FoodCourt, ShoppingMall, Spa, 
# ----------------------
# 12- VRDeck - Amount the passenger has billed at each of the Spaceship Titanic's many luxury amenities.
# ----------------------
# 13- Name - The first and last names of the passenger.
# ----------------------
# 14- Transported - Whether the passenger was transported to another dimension.
#  This is the target, the column you are trying to predict.
```

```python
#import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split #pip install scikit-learn si no lo tenemos instalado
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head(10)
```

**Check the shape of your data**

```python
spaceship.shape
```

**Check for data types**

```python
spaceship.dtypes
```

**Check for missing values**

```python
spaceship.isnull().sum()
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
spaceship = spaceship.dropna()
spaceship.shape
```

**KNN**

K Nearest Neighbors is a distance based algorithm, and requeries all **input data to be numerical.**

Let's only select numerical columns as our features.

```python
features  = spaceship.select_dtypes(include=["int64", "float64"])  # Variables de entrada al modelo
```

And also lets define our target.

```python
target = spaceship["Transported"]
```

**Train Test Split**

Now that we have split the data into **features** and **target** variables and imported the **train_test_split** function, split X and y into X_train, X_test, y_train, and y_test. 80% of the data should be in the training set and 20% in the test set.

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.20, random_state=0)
```

| Variable | Qué contiene | Para qué sirve |
| --- | --- | --- |
| **X_train** | Features (80%) | Entrenar el modelo |
| **X_test** | Features (20%) | Evaluar el modelo |
| **y_train** | Target (80%) | Entrenar |
| **y_test** | Target (20%) | Evaluar |

```python
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)

```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

You need to choose between **Classificator** or **Regressor**.  
Take into consideration target variable to decide.

Initialize a KNN instance without setting any hyperparameter.

```python
from sklearn.neighbors import KNeighborsClassifier
```

Fit the model to your data.

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

Evaluate your model.

```python
knn
```

| Parámetro | Valor | Qué significa | Impacto en el modelo |
| --- | --- | --- | --- |
| **n_neighbors** | 3 | Número de vecinos que se usan para clasificar | Muy alto |
| **weights** | 'uniform' | Todos los vecinos pesan igual | Medio |


```python
knn.fit(X_train, y_train)
```

Hago las predicciones

| Variable | Qué contiene | Para qué sirve |
| --- | --- | --- |
| **X_train** | Features (80%) | Entrenar el modelo |
| **X_test** | Features (20%) | Evaluar el modelo |
| **y_train** | Target (80%) | Entrenar |
| **y_test** | Target (20%) | Evaluar |

```python
pred = knn.predict(X_test) # esta es la prediccion para usar luego
pred
```

Vamos a sacar los valores reales que deberia dar para comparar

```python
y_test.values
```


Para poder evaluar nuestro modelo utilizaremos el método "score" que nos dará precisión

```python
knn.score(X_test, y_test) # accuracy
```

| Variable | Qué contiene | Para qué sirve |
| --- | --- | --- |
| **X_train** | Features (80%) | Entrenar el modelo |
| **X_test** | Features (20%) | Evaluar el modelo |
| **y_train** | Target (80%) | Entrenar |
| **y_test** | Target (20%) | Evaluar |

Accuracy es siempre global.
Precision, recall y F1 son por clase primero, y globales solo después de promediarlas.

Ahora comparamos la prediccion con el resultado segun diferentes funciones

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
# accuracy
print(accuracy_score(y_test, pred)) #le paso las predicciones manualmente. Sale lo mismo que con knn.score(X_test, y_test)

# precision
print(precision_score(y_test, pred, average='macro'))

# recall
print(recall_score(y_test, pred, average='macro'))

# F1-score
print(f1_score(y_test, pred, average='macro'))
```

Vamos a ver que dice la confusion matrix

```python
# confusion matrix

##[[TN FP]
 ##[FN TP]]

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, pred)
```

```python
# scikit‑learn ordena las clases en orden alfabético / lógico antes de construir la matriz de confusión.
# por tanto False = 0 y true = 1  y sale asi la matriz 
#  [[TN, FP],     -----------   
#  [FN, TP]]

# TN = 477 TN (True Negatives)  : aciertos en NO transportados
# FP = 184 FP (False Positives) : dice transportado pero no lo es
# FN = 166 FN (False Negatives) : dice no transportado pero sí lo es
# TP = 495 TP (True Positives) : aciertos en transportados

```

```python
# Pedimos a scikit‑learn que te imprima un informe completo de cómo ha clasificado tu modelo, mucho más detallado que accuracy.

# La columna support solo cuenta pasajeros reales en cada tipo 
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred))
```

CONCLUSION:
